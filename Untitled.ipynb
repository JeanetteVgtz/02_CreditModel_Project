```python
# ---------------------------------------------------------------
# CREDIT MODEL PROJECT - PHASE 1: DATA PREPARATION
# ---------------------------------------------------------------

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid", palette="crest")

# Cargar datos
df = pd.read_csv("train.csv")

print("Dimensiones:", df.shape)
print("\nTipos de datos:")
print(df.dtypes)

# Eliminar columnas irrelevantes para scoring
drop_cols = ["ID", "Customer_ID", "Name", "SSN", "Type_of_Loan", "Month"]
df.drop(columns=drop_cols, inplace=True, errors="ignore")

# --- 1. Convertir variable objetivo a numérica ---
df["Credit_Score"] = df["Credit_Score"].replace({"Poor": 0, "Standard": 1, "Good": 2})

# --- 3. Codificar variables categóricas ---
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    df[col] = le.fit_transform(df[col])

print("\n Data:")
print(df.head(3))
```

```python
# ---------------------------------------------------------------
# Exploración visual
# ---------------------------------------------------------------

# Distribución de la variable objetivo
sns.countplot(x="Credit_Score", data=df, palette="crest")
plt.title("Distribución de los Niveles de Credit Score")
plt.show()
```

```python
# Edad vs Score
sns.boxplot(x="Credit_Score", y="Age", data=df, palette="coolwarm")
plt.title("Distribución de Edad por Nivel de Score")
plt.show()
```

```python
# Ingresos vs Score
sns.boxplot(x="Credit_Score", y="Annual_Income", data=df, palette="crest")
plt.title("Ingresos Anuales por Nivel de Score")
plt.show()

```

```python
# Histograma de tasas de interés
sns.histplot(df["Interest_Rate"], bins=30, kde=True, color="teal")
plt.title("Distribución de Tasas de Interés")
plt.show()
```

```python
# Correlaciones numéricas
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), cmap="viridis", center=0)
plt.title("Mapa de Correlación - Variables Numéricas")
plt.show()
```

```python
# ---------------------------------------------------------------
# CREDIT MODEL PROJECT - MODEL TRAINING & COMPARISON
# ---------------------------------------------------------------
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, roc_auc_score, classification_report,
    confusion_matrix, roc_curve
)


# ---------------------------------------------------------------
# Definir variables predictoras (X) y variable objetivo (y)
# ---------------------------------------------------------------
X = df.drop(columns=["Credit_Score"])
y = df["Credit_Score"]

# Si Credit_Score tiene más de 2 clases (Poor/Standard/Good),
# lo convertimos a binario: 0 = bajo riesgo, 1 = bueno o estándar
if len(y.unique()) > 2:
    y = np.where(y >= 1, 1, 0)

# Codificar categóricas si aún existen (por si no estaban numéricas)
X = X.apply(lambda col: LabelEncoder().fit_transform(col) if col.dtype == 'object' else col)

# Escalar datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir en train/test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# ---------------------------------------------------------------
# REGRESIÓN LOGÍSTICA
# ---------------------------------------------------------------
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)

y_pred_log = log_model.predict(X_test)
y_pred_log_prob = log_model.predict_proba(X_test)[:,1]

acc_log = accuracy_score(y_test, y_pred_log)
auc_log = roc_auc_score(y_test, y_pred_log_prob)

print("\n---- LOGISTIC REGRESSION ----")
print(f"Train Accuracy: {log_model.score(X_train, y_train):.3f}")
print(f"Test Accuracy : {acc_log:.3f}")
print(f"Test AUC      : {auc_log:.3f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_log))
```

```python
# ---------------------------------------------------------------
#  RANDOM FOREST
# ---------------------------------------------------------------
rf_model = RandomForestClassifier(
    n_estimators=300, 
    max_depth=10,
    random_state=42
)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)
y_pred_rf_prob = rf_model.predict_proba(X_test)[:,1]

acc_rf = accuracy_score(y_test, y_pred_rf)
auc_rf = roc_auc_score(y_test, y_pred_rf_prob)

print("\n---- RANDOM FOREST ----")
print(f"Train Accuracy: {rf_model.score(X_train, y_train):.3f}")
print(f"Test Accuracy : {acc_rf:.3f}")
print(f"Test AUC      : {auc_rf:.3f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))
```

```python
# ---------------------------------------------------------------
# MATRICES DE CONFUSIÓN
# ---------------------------------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(12,5))
sns.heatmap(confusion_matrix(y_test, y_pred_log), annot=True, fmt="d", cmap="Blues", ax=axes[0])
axes[0].set_title("Confusion Matrix - Logistic Regression")
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt="d", cmap="Greens", ax=axes[1])
axes[1].set_title("Confusion Matrix - Random Forest")
plt.show()
```

```python
# ---------------------------------------------------------------
# CURVAS ROC
# ---------------------------------------------------------------
fpr_log, tpr_log, _ = roc_curve(y_test, y_pred_log_prob)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_pred_rf_prob)

plt.figure(figsize=(8,6))
plt.plot(fpr_log, tpr_log, label=f"Logistic Regression (AUC={auc_log:.3f})")
plt.plot(fpr_rf, tpr_rf, label=f"Random Forest (AUC={auc_rf:.3f})")
plt.plot([0,1], [0,1], 'k--', label="Random Classifier (AUC=0.5)")
plt.title("ROC Curve Comparison")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

```python
# ---------------------------------------------------------------
# FEATURE IMPORTANCE - RANDOM FOREST
# ---------------------------------------------------------------
importances = pd.Series(rf_model.feature_importances_, index=X.columns).sort_values(ascending=True)
plt.figure(figsize=(8,6))
importances.tail(15).plot(kind='barh', color='teal')
plt.title("Top 15 Most Important Features (Random Forest)")
plt.xlabel("Importance")
plt.show()
```

```python
# ---------------------------------------------------------------
# CREDIT MODEL PROJECT - MODEL OPTIMIZATION & VALIDATION
# ---------------------------------------------------------------
from sklearn.model_selection import cross_val_score, GridSearchCV, StratifiedKFold
print(" Validación cruzada y ajuste de hiperparámetros\n")

# ---------------------------------------------------------------
# VALIDACIÓN CRUZADA (K-Fold)
# ---------------------------------------------------------------
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_scores_log = cross_val_score(log_model, X_scaled, y, cv=cv, scoring="roc_auc")
cv_scores_rf  = cross_val_score(rf_model,  X_scaled, y, cv=cv, scoring="roc_auc")

print(f"Regresión Logística - AUC promedio: {cv_scores_log.mean():.4f} ± {cv_scores_log.std():.4f}")
print(f"Random Forest - AUC promedio:       {cv_scores_rf.mean():.4f} ± {cv_scores_rf.std():.4f}")

# ---------------------------------------------------------------
# AJUSTE DE HIPERPARÁMETROS (GridSearchCV)
# ---------------------------------------------------------------
param_grid = {
    "n_estimators": [100, 300, 500],
    "max_depth": [5, 10, 15, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    scoring="roc_auc",
    cv=cv,
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train, y_train)

print("\n Mejor combinación de hiperparámetros:")
print(grid_search.best_params_)

best_rf = grid_search.best_estimator_
```

```python
# ---------------------------------------------------------------
# REENTRENAR MODELO ÓPTIMO Y EVALUAR
# ---------------------------------------------------------------
y_pred_best = best_rf.predict(X_test)
y_pred_best_prob = best_rf.predict_proba(X_test)[:,1]

acc_best = accuracy_score(y_test, y_pred_best)
auc_best = roc_auc_score(y_test, y_pred_best_prob)

print("\n---- RANDOM FOREST OPTIMIZADO ----")
print(f"Train Accuracy: {best_rf.score(X_train, y_train):.3f}")
print(f"Test Accuracy : {acc_best:.3f}")
print(f"Test AUC      : {auc_best:.3f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_best))
```

```python
# ---------------------------------------------------------------
# COMPARACIÓN VISUAL DE ROC (ANTES VS DESPUÉS)
# ---------------------------------------------------------------
fpr_base, tpr_base, _ = roc_curve(y_test, y_pred_rf_prob)
fpr_best, tpr_best, _ = roc_curve(y_test, y_pred_best_prob)

plt.figure(figsize=(8,6))
plt.plot(fpr_base, tpr_base, label=f"Random Forest Base (AUC={auc_rf:.3f})", linestyle="--")
plt.plot(fpr_best, tpr_best, label=f"Optimized Random Forest (AUC={auc_best:.3f})", color="green")
plt.plot([0,1],[0,1],'k--',label="Random Classifier (AUC=0.5)")
plt.title("ROC Curve: Base vs Optimized Model")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```
